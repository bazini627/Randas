```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import datetime
```

### Read in the CA Counties GeoJSON

```python
ca_counties = gpd.read_file('../data/caCountiesNoChannelIslands.geojson')
```

### Quick plot of the GeoJSON

```python
ca_counties.plot()
```

### Drop a few of the columns in ca_counties since we won't really need them for this basic map

```python
ca_counties = ca_counties.drop(['COUNTYNS', 'GEOID','ALAND','AWATER'], axis=1)
```

### Lowercase our column headers for ca_counties

```python
ca_counties = ca_counties.rename(str.lower, axis=1)
```

### Add a coords column to ca_counties which will be used to apply labels to the map polygons

```python
# Use Shapely's representative_point on the geometry column to create a coords column which will be used to display the county name
# It's a roughly calculated point that will fall within the county polygon
ca_counties['coords'] = ca_counties['geometry'].apply(lambda x: x.representative_point().coords[:])
ca_counties.head()
```

### Coords column is a list so just set the column value to be the first value of the list

```python
ca_counties['coords'] = [coords[0] for coords in ca_counties['coords']]
```

```python
ca_counties.head()
```

### Set county(name) as the index

```python
ca_counties.set_index('name', inplace=True)
```

### Function to add county labels to maps

```python
def add_labels(df):
    for index, row in df.iterrows():
        if index == 'San Francisco' or index == 'San Mateo':
            plt.annotate(s=index, xy=row['coords'],horizontalalignment='right',fontsize=8.25)
        elif index == 'Placer' or index == 'Yuba':
            plt.annotate(s=index, xy=row['coords'],horizontalalignment='left',fontsize=8.25)
        elif index == 'Stanislaus':
             plt.annotate(s=index, xy=row['coords'],horizontalalignment='center', xytext=(6,6), textcoords='offset points', fontsize=8.25)
        elif index == 'Los Angeles' or index == 'Kern' or index == 'Fresno':
             plt.annotate(s=index, xy=row['coords'], xytext=(9,9), textcoords='offset points', fontsize=8.25)
        else:
            plt.annotate(s=index, xy=row['coords'],horizontalalignment='center', verticalalignment='center_baseline', fontsize=8.25)
```

### Create our map

```python
# Base map of CA
basemap = ca_counties.plot(figsize=(10,10), color='white', edgecolor='black', alpha=0.5)

# Turn off axis
basemap.set_axis_off()

# Title For Map
plt.title('California Counties' , fontsize=20);

# Add the county labels
add_labels(ca_counties)
```
